1. Create a dataset with 10,000 rows and 4 random variables: 2 of them normally distributed, 2 uniformly distributed

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#normally distributed variables
n = 10000
x1 = np.random.normal(3,2,size=n)
x2 = np.random.normal(5,1, size=n)
```

```python
#uniformly distributed variables
x3 = np.random.uniform(size=n)
x4 = np.random.uniform(size=n)
```

2. Add another variable ("y") as a linear combination, with some coefficients of your choice, of: the 4 variables above; the square of one of the variables; and some random "noise" (randomly distributed, centered at 0, small variance). Hint: think about y = a1*x1+a2*x2+a3*x3+a4*x4+epsilon. You will get a dataset with 5 variables.

```python
#random noise
epsilon = np.random.normal(0, 1, size = n)

#coefficients
a1 = 0.2
a2 = 0.3
a3 = 0.8
a4 = 2
a5 = 1.2

y = a1*x1 + a2*x2 + a3*x3 + a4*x4 + a5*(x1**2) + epsilon
```

```python
#creating dataframe
df = pd.DataFrame({
    "x1": x1, "x2": x2, "x3": x3, "x4": x4, 'x1_squared':x1**2 , "y": y,
})
```

```python
df
```

3. Split the dataset in #2 into 70% for training and 30% for testing

```python
#splitting X and y into separate dataframes
X = df[['x1', 'x2', 'x3', 'x4', 'x1_squared']]
y = df['y']

#creating a 70/30 split for training and testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0)
```

4. Estimate the linear regression coefficients using OLS for the training data; compute the Mean Standard Error (average of squared differences between estimates and actuals) on both the training dataset, and the testing dataset

```python
#creating a model to fit my training set
ols = LinearRegression(fit_intercept=True)
ols.fit(X_train, y_train)
```

```python
#calculating MSE for training and testing set
train_mse = mean_squared_error(y_train, ols.predict(X_train))
test_mse = mean_squared_error(y_test, ols.predict(X_test))

print(f"Mean Squared Error (Training): {train_mse:.4f}")
print(f"Mean Squared Error (Testing): {test_mse:.4f}")

#finding coefficients and intercept of model
intercept = np.round(ols.intercept_, 4)
coeff = np.round(ols.coef_, 4)
print("Intercept:", intercept, "Coefficients:", coeff)
```

The MSE for the training and testing datasets are very small which suggests that our model fits our data pretty well. In addition, our MSE score for the training and testing dataset are very close which can suggest our model is not overfitting our data.

5. Use bootstrapping to create 10 other samples from the data your created in #2 above

```python
#bootstrapping to create 10 new samples with replacement
bootstrapping = []
n_samples = 10
for i in range(n_samples):
    sample = df.sample(n=len(df), replace=True, random_state=i)
    bootstrapping.append(sample)
```

6. Estimate the linear regression coefficients using OLS for each of the 10 bootstrap samples in #5

```python
#each sample will now be fit our model
sample_coeff = []
for sample_df in bootstrapping:
    X_boot = sample_df[['x1', 'x2', 'x3', 'x4', 'x1_squared']]
    y_boot = sample_df['y']
    
    ols_boot = LinearRegression(fit_intercept=True)
    ols_boot.fit(X_boot, y_boot)
    
    intercept_and_coeff = [ols_boot.intercept_] + list(ols_boot.coef_)
    sample_coeff.append(intercept_and_coeff)
sample_coeff
```

7. For each linear regression parameter, use the estimates computed in #6 to get the mean and the standard deviation.

```python
#using estimates from our bootstrapping to get the mean and standard deviation for each parameter
bootstrap_coeffs_array = np.array(sample_coeff)
mean = np.mean(bootstrap_coeffs_array, axis=0)
standard_deviation = np.std(bootstrap_coeffs_array, axis=0)

results = pd.DataFrame({
    'Parameter': ['Intercept', 'x1', 'x2', 'x3', 'x4', 'x1^2'],
    'Bootstrap Mean': mean,
    'Bootstrap Std Dev': standard_deviation
})
results
```

8. What can you say about the coefficients in #4 looking at the results in #7?

Our mean estimates from our bootstrapping are close to the estimates seen in #4 (our coeffiecients) which further indicates that our estimates are consistent. The standard deviation of our parameters are all close to zero which shows us that our coefficients are precise and significant. Overall, our ols model was able to fit out data very well and the bootstrapping results confirm that the coefficients are stable.
