```python
import os
import json
from pathlib import Path

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error
import joblib
```

```python
%run Clean_data_JST_recovery.ipynb
```

```python
RAW_DIR = ROOT / 'Raw Data'
JST_PATH = RAW_DIR / 'JSTdatasetR6.xlsx'

df = pd.read_excel(JST_PATH)
print('Raw JST shape:', df.shape)
df.head()
```

```python
X, y, training_columns, continuous_cols, feature_means = clean_data(
    df,
    is_training=True,
    verbose=True,
    cfg=CONFIG
)

print('Episode feature matrix shape:', X.shape)
print('Target length:', len(y))
X.head()
```

```python
# cleaned panel + engineered features
panel = basic_panel_clean(df, CONFIG)
panel = add_engineered_features(panel, CONFIG)
onsets = find_crisis_onsets(panel, CONFIG)


c = CONFIG['country_col']
ycol = CONFIG['year_col']
gdp_col = CONFIG['gdp_col']

valid_mask = []
for _, row in onsets.iterrows():
    country = row[c]
    t0 = int(row['crisis_year'])
    cdf = panel[panel[c] == country]
    rt = compute_recovery_time(
        cdf, t0, gdp_col=gdp_col, year_col=ycol,
        pre_peak_window=int(CONFIG['pre_peak_window']),
        max_horizon=int(CONFIG['max_horizon'])
    )
    valid_mask.append(pd.notna(rt))

onsets_valid = onsets.loc[valid_mask].reset_index(drop=True)

print('Onsets:', len(onsets), 'Valid episodes:', len(onsets_valid))
onsets_valid.head()
```

```python
train_end = 1980
val_end = 2000

crisis_year = onsets_valid['crisis_year'].values

train_idx = np.where(crisis_year <= train_end)[0]
val_idx = np.where((crisis_year > train_end) & (crisis_year <= val_end))[0]
test_idx = np.where(crisis_year > val_end)[0]

print('Train episodes:', len(train_idx))
print('Val episodes:', len(val_idx))
print('Test episodes:', len(test_idx))

X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
X_val, y_val = X.iloc[val_idx], y.iloc[val_idx]
X_test, y_test = X.iloc[test_idx], y.iloc[test_idx]
```

```python
OUT_DIR = ROOT / 'output'
FIG_DIR = OUT_DIR / 'figures'
MODEL_DIR = ROOT / 'model'

OUT_DIR.mkdir(exist_ok=True, parents=True)
FIG_DIR.mkdir(exist_ok=True, parents=True)
MODEL_DIR.mkdir(exist_ok=True, parents=True)

plt.figure(figsize=(8, 5))
sns.histplot(y, bins=15)
plt.title('Recovery Time (years) Distribution')
plt.tight_layout()
plt.savefig(FIG_DIR / 'recovery_time_hist.png', dpi=200)
plt.show()

plt.figure(figsize=(10, 8))
corr = X.corr()
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title('Feature Correlations (Episode-level)')
plt.tight_layout()
plt.savefig(FIG_DIR / 'feature_corr_heatmap.png', dpi=200)
plt.show()
```

```python
numeric_features = training_columns  

linear_preprocess = ColumnTransformer(
    transformers=[
        ('num', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', StandardScaler())
        ]), numeric_features)
    ],
    remainder='drop'
)

rf_preprocess = ColumnTransformer(
    transformers=[
        ('num', Pipeline([
            ('imputer', SimpleImputer(strategy='median'))
        ]), numeric_features)
    ],
    remainder='drop'
)

# Model objects
ridge_pipe = Pipeline([
    ('prep', linear_preprocess),
    ('model', Ridge())
])

lasso_pipe = Pipeline([
    ('prep', linear_preprocess),
    ('model', Lasso(max_iter=20000))
])

rf_pipe = Pipeline([
    ('prep', rf_preprocess),
    ('model', RandomForestRegressor(random_state=42))
])
```

```python
scoring = 'neg_mean_absolute_error'

ridge_grid = GridSearchCV(
    ridge_pipe,
    param_grid={'model__alpha': [0.01, 0.1, 1, 10, 50, 100]},
    scoring=scoring,
    cv=5,
    n_jobs=-1
)
ridge_grid.fit(X_train, y_train)

lasso_grid = GridSearchCV(
    lasso_pipe,
    param_grid={'model__alpha': [0.001, 0.01, 0.1, 1, 5, 10]},
    scoring=scoring,
    cv=5,
    n_jobs=-1
)
lasso_grid.fit(X_train, y_train)

rf_grid = GridSearchCV(
    rf_pipe,
    param_grid={
        'model__n_estimators': [200, 500],
        'model__max_depth': [None, 5, 10],
        'model__min_samples_leaf': [1, 2, 5]
    },
    scoring=scoring,
    cv=5,
    n_jobs=-1
)
rf_grid.fit(X_train, y_train)

print('Best Ridge params:', ridge_grid.best_params_, 'CV MAE:', -ridge_grid.best_score_)
print('Best Lasso params:', lasso_grid.best_params_, 'CV MAE:', -lasso_grid.best_score_)
print('Best RF params:', rf_grid.best_params_, 'CV MAE:', -rf_grid.best_score_)
```

```python
def evaluate(model, X, y_true, name='model'):
    pred = model.predict(X)
    mae = mean_absolute_error(y_true, pred)
    rmse = mean_squared_error(y_true, pred, squared=False)
    return {'name': name, 'mae': mae, 'rmse': rmse}

best_ridge = ridge_grid.best_estimator_
best_lasso = lasso_grid.best_estimator_
best_rf = rf_grid.best_estimator_

val_results = []
if len(X_val) > 0:
    val_results.append(evaluate(best_ridge, X_val, y_val, 'Ridge'))
    val_results.append(evaluate(best_lasso, X_val, y_val, 'Lasso'))
    val_results.append(evaluate(best_rf, X_val, y_val, 'RandomForest'))
else:
    print('No validation set (val_idx is empty). Adjust split years to create a validation set.')

val_results_df = pd.DataFrame(val_results).sort_values('mae')
val_results_df
```

```python
if len(val_results_df) == 0:
    raise RuntimeError('No validation results to choose from. Adjust split years to ensure val set exists.')

final_name = val_results_df.iloc[0]['name']
final_model = {'Ridge': best_ridge, 'Lasso': best_lasso, 'RandomForest': best_rf}[final_name]

print('Selected final model:', final_name)

test_metrics = evaluate(final_model, X_test, y_test, f'{final_name} (TEST)') if len(X_test) > 0 else None

# Baseline
baseline_pred = np.full(len(y_test), y_train.mean())
baseline_mae = mean_absolute_error(y_test, baseline_pred)
baseline_rmse = mean_squared_error(y_test, baseline_pred, squared=False)

print('TEST metrics:', test_metrics)
print('Baseline TEST MAE:', baseline_mae, 'RMSE:', baseline_rmse)
```

```python
# Save model
joblib.dump(final_model, MODEL_DIR / 'recovery_model.joblib')

# Save metadata
(MODEL_DIR / 'training_columns.json').write_text(json.dumps(training_columns, indent=2))
(MODEL_DIR / 'feature_means.json').write_text(json.dumps(feature_means, indent=2))
(MODEL_DIR / 'config_snapshot.json').write_text(json.dumps(CONFIG, indent=2))

# Save results summary
results = {
    'final_model': final_name,
    'val_results': val_results_df.to_dict(orient='records'),
    'test_metrics': test_metrics,
    'baseline_test_mae': float(baseline_mae),
    'baseline_test_rmse': float(baseline_rmse),
}
(OUT_DIR / 'results.json').write_text(json.dumps(results, indent=2))

print('Saved model to:', MODEL_DIR / 'recovery_model.joblib')
print('Saved metadata + results to:', MODEL_DIR, 'and', OUT_DIR)
```
