```python
print(df_train['gender'].unique())
print(df_train['hypertension'].unique())
print(df_train['heart_disease'].unique())
print(df_train['ever_married'].unique())
print(df_train['work_type'].unique())
print(df_train['Residence_type'].unique())
print(df_train['smoking_status'].unique())
```

```python
df_train['gender'] = df_train['gender'].map({'Male': 0, 'Female': 1, 'Other': 2})
df_train['ever_married'] = df_train['ever_married'].map({'No': 0, 'Yes': 1})
df_train['Residence_type'] = df_train['Residence_type'].map({'Rural': 0, 'Urban': 1})
df_train = pd.get_dummies(df_train, columns=['smoking_status'], drop_first=True)
df_train = pd.get_dummies(df_train, columns=['work_type'], drop_first=True)
```

```python
df_train.drop(columns=['id'], inplace=True)
```

```python
df_train.to_csv('../ProcessedData/stroke_cleaned.csv', index=False)
```
